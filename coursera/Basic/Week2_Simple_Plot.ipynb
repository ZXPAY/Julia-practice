# Simple plot with the Plots package

* save the figure to *.png or *.pdf

```julia
using Plots    # need wait a second
```

```julia
using DelimitedFiles
start_index = 3
raw_data = DelimitedFiles.readdlm("Ebola_epidaysConverted.csv", ',')
```

```julia
gr()
```

```julia
epidays = raw_data[start_index:end, 1]
case1 = raw_data[start_index:end, 2]
case2 = raw_data[start_index:end, 4]
case3 = raw_data[start_index:end, 6]
for i=1:length(case1)
    ss = "0"
    for s in string(case1[i])
        if s != ',' && isnumeric(s)
            ss *= s
        end
    end
    case1[i] = parse(Float32, ss)
    ss = "0"
    for s in string(case2[i])
        if s != ',' && isnumeric(s)
            ss *= s
        end
    end
    case2[i] = parse(Float32, ss)
    ss = "0"
    for s in string(case3[i])
        if s != ',' && isnumeric(s)
            ss *= s
        end
    end
    case3[i] = parse(Float32, ss)
end
```

```julia
plot(epidays, case1, linetype=:scatter, marker=:diamond)
```

```julia
plot(epidays, case1, 
    title="West African EVD epidemic, total cases",
    xlabel="day",
    ylabel="death",
    marker=(:diamond, 8, "gray"),
    line=(:path),
    legend=false,
    grid=true
    )
```

```julia
y = [case1 case2 case3]
```

```julia
plot(epidays, y,
    legend=:topleft,
    title="this is title",
    label=["case1" "case2" "case3"],
    marker=([:octagon :star7 :square], 5),
    xlabel="days",
    ylabel="death",
    line=(:scatter, 8)
    )
```

```julia
savefig("savefig.pdf")
```
